```python
import math
import numpy
import matplotlib.pyplot as plt
from numpy import genfromtxt
outData = genfromtxt("D:\\Julia\\data\\dubinsOutCirc.csv", delimiter=',')
inData = genfromtxt("D:\\Julia\\data\\dubinsInCirc.csv", delimiter=',')
```

# Preprocessing outside datapoints

```python
outData[:,-1] = outData[:,-1].round()

outData[:,0] = outData[:,0] - outData[:,2]
outData[:,1] = outData[:,1] - outData[:,3]

outData[:,2] = 0
outData[:,3] = 0

tempX = numpy.multiply(outData[:,0], numpy.cos(outData[:,4])) + numpy.multiply(outData[:,1], numpy.sin(outData[:,4]))
tempY = -numpy.multiply(outData[:,0], numpy.sin(outData[:,4])) + numpy.multiply(outData[:,1], numpy.cos(outData[:,4]))

outData[:,0] = tempX
outData[:,1] = tempY

outData[:,4] = 0
print(outData.shape)

# REDUCING u=0 POINTS (choosing random half of them)

ys = outData[:,1].reshape(1,len(outData))[0]

temp = numpy.arange(len(ys))
a = (ys<-0.055)|(ys>0.055)
b = temp%2000==0
outData = outData[a | b, :]

us = outData[:,5].reshape(1,len(outData))[0]
c = us!=0
outData = outData[c,:]


print(outData.shape)
```

# SIMULATION

```python
def normalizer(data):
    data[:,0] = data[:,0] - data[:,2]
    data[:,1] = data[:,1] - data[:,3]
    
    data[:,2] = 0
    data[:,3] = 0
    
    tempX = numpy.multiply(data[:,0], numpy.cos(data[:,4])) + numpy.multiply(data[:,1], numpy.sin(data[:,4]))
    tempY = -numpy.multiply(data[:,0], numpy.sin(data[:,4])) + numpy.multiply(data[:,1], numpy.cos(data[:,4]))
    
    data[:,0] = tempX
    data[:,1] = tempY
    
    data[:,4] = 0
    
    data = data[:,[0,1]]
    
    return data
```

```python
def inDubinsCircle(x,y):
    value = (x**2 + (y-1)**2 - 1)*(x**2 + (y+1)**2 - 1)
    return value<0
```

```python
# defining initial conditions
xi = 0
yi = 0
xf = 0.5
yf = 0.5
ai = 0
v = 1
```

```python
x = [xi]
y = [yi]
a = [ai]
dt = 0.01
us = []

if inDubinsCircle(xf,yf):
    classifier = inClf
    print("inClf used")
else:
    classifier = outClf
    print("outClf used")

while math.sqrt((xf - x[-1])**2 + (yf - y[-1])**2) > 0.05:
    sample = numpy.array([xf,yf,x[-1],y[-1],a[-1]]).reshape(1, -1)
    sample = normalizer(sample)
    
    u = classifier.predict(sample)
    us.append(u)
    a.append(a[-1] + dt*v*u[0])
    x.append(x[-1] + dt*v*math.cos(a[-2]))
    y.append(y[-1] + dt*v*math.sin(a[-2]))
```

```python
plt.plot(x, y)
plt.grid()
plt.arrow(x[len(x)//2], y[len(y)//2],x[len(x)//2+3]-x[len(x)//2] ,y[len(y)//2+3]-y[len(y)//2] , shape='full', lw=0, length_includes_head=True, head_width=.3)

plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')

circle1 = plt.Circle((0, 1), 1, color='y')
circle2 = plt.Circle((0, -1), 1, color='y')

axes=plt.gca()
axes.add_patch(circle1)
axes.add_patch(circle2)
axes.set_aspect('equal')
#axes.set(title="x_f = " + str(xf) + ", y_f = " + str(yf))
plt.savefig("D:\\Julia\\data\\newImages\\inClf.pdf",pad_inches=0.1,bbox_inches='tight')
```
